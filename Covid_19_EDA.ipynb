Visualization of Dataset

```python
# Example DataFrame df with relevant columns
df = pd.DataFrame({
    'Confirmed': [100, 200, 300],
    'Deaths': [10, 20, 30],
    'Recovered': [50, 100, 150],
    'Active': [40, 80, 120]
})

# Define the list of column names
status_count = ['Confirmed', 'Deaths', 'Recovered', 'Active']

# Melt the DataFrame to convert it into long format for easier plotting
df_melted = df.melt(var_name='Status', value_name='Count')

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Count', y='Status', hue='Status', dodge=False)
plt.xlabel("Count")
plt.ylabel("Status")
plt.title("Distribution of Status Categories")
plt.legend(title=None, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# Calculate global active cases and total cases
global_active_cases = df['Active'].sum()
global_cases = df[['Confirmed', 'Deaths', 'Recovered']].sum().sum()

# Create a DataFrame for plotting
data = {
    'Cases': ['Active Cases', 'Other Cases'],
    'Count': [global_active_cases, global_cases - global_active_cases]
}
df_plot = pd.DataFrame(data)

# Plotting the countplot
plt.figure(figsize=(8, 6))
sns.barplot(data=df_plot, x='Cases', y='Count')
plt.title('Global Active Cases Distribution')
plt.xlabel('Cases')
plt.ylabel('Count')
```
